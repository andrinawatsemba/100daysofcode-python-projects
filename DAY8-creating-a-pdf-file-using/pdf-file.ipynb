```python
from reportlab.pdfgen import canvas

generated_pdf = canvas.Canvas("Andrina'spdf.pdf")

generated_pdf.drawString(100, 750, "welcome to Andrina's pdf file")
generated_pdf.drawString(100, 735,  "this is a simple file in python")
generated_pdf.drawString(100, 720, "generated using reportlab library")
generated_pdf.drawString(100, 705, " I am out here testing random python scripts/projects, anything to put my knowledge of the python syntax to use")
generated_pdf.drawString(100, 690, "okay bye")

generated_pdf.save()
```
